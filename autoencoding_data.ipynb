```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Load data
data = pd.read_csv('sensor.csv')

# Normalize data
data_numerical = data.select_dtypes(include=[np.number])  # This selects only numerical columns

# Normalize data
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data_numerical.values)
print(data_normalized)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, RepeatVector, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping

# Define model architecture
data_numerical = data.select_dtypes(include=[np.number])
data_normalized = scaler.fit_transform(data_numerical.values)

# Reshape data for LSTM layer
# Assuming 'data_normalized' is already defined and normalized as shown in your notebook

# Slice the first 3000 values for training
data_normalized_subset = data_normalized[:3000]

# Reshape data for LSTM layer
data_reshaped = data_normalized_subset.reshape((data_normalized_subset.shape[0], data_normalized_subset.shape[1], 1))

# Define the model architecture (assuming the model definition is the same)
model = Sequential([
    LSTM(64, activation='relu', input_shape=(data_normalized_subset.shape[1], 1), return_sequences=True),
    LSTM(32, activation='relu', return_sequences=False),
    RepeatVector(data_normalized_subset.shape[1]),
    LSTM(32, activation='relu', return_sequences=True),
    LSTM(64, activation='relu', return_sequences=True),
    TimeDistributed(Dense(1))
])

model.compile(optimizer='adam', loss='mse')

# Train the model with optimizations
model.fit(data_reshaped, data_reshaped, epochs=20, batch_size=128, validation_split=0.2)
```

```python
# Predict on the same dataset
reconstructed = model.predict(data_reshaped)

# Calculate MSE as reconstruction error
reconstruction_error = tf.keras.losses.mse(data_reshaped, reconstructed)

# Detect anomalies (customize threshold as needed)
threshold = 0.01  # This is a hypothetical threshold
anomalies = reconstruction_error > threshold
```

```python
import matplotlib.pyplot as plt

# Assuming 'reconstruction_error' is a numpy array or a similar structure
plt.figure(figsize=(10, 6))
plt.plot(reconstruction_error, label='Reconstruction Error')
plt.axhline(y=threshold, color='r', linestyle='--', label='Threshold')
plt.title('Reconstruction Error over Time')
plt.xlabel('Time Index')
plt.ylabel('Error')
plt.legend()
plt.show()
```

```python
# Find indices of anomalies
anomaly_indices = [i for i, error in enumerate(reconstruction_error) if error > threshold]

# Print anomaly details
print("Anomaly Indices:", anomaly_indices)
print("Anomaly Values:", data.iloc[anomaly_indices])
```

```python
context_window = 5  # Number of data points before and after the anomaly

for index in anomaly_indices:
    start_index = max(0, index - context_window)
    end_index = min(len(data), index + context_window)
    print("Anomaly at Index:", index)
    print(data.iloc[start_index:end_index])
```
